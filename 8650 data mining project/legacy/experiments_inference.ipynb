```python
import torch
import argparse
import classifier

device = torch.device('cuda:0') if torch.cuda.is_available() else torch.device('cpu')

PATH = 'deface_model.pt'
def load_model(PATH):
    state_dict = torch.load(PATH, map_location="cpu", weights_only=False)
    args = argparse.Namespace(**{**vars(state_dict["args"])})    
    model = classifier.Classifier(args) 
    model.load_state_dict(state_dict["model"])
    return model


model = load_model(PATH)
model = model.to(device) 
```

```python
#import dataload
import nibabel as nib
import skimage.transform as skTrans
import numpy as np


filepath = filepath + '.gz' if not filepath.endswith('.gz') else filepath 

# print(filepath)
img = nib.load(filepath)
print(img.get_fdata().shape)
res = skTrans.resize(img.get_fdata(), (256,256,150), order=1, preserve_range=True)
resized_img = nib.Nifti1Image(res, img.affine).get_fdata()

torch_tensor = torch.tensor(np.nan_to_num(resized_img), dtype=torch.float32)
torch_tensor = torch_tensor.unsqueeze(0)
out = model(torch_tensor.cuda())
print(out)
```
